```python
data_directory = '/Users/lib278/Data/PRIMORDIAL_DISKS/K2_TESS/'
lightcurve_directory = data_directory + 'Lightcurves/'
periodogram_directory = data_directory + 'Periodograms/'
```

```python
lightcurve_K2_files = []
lightcurve_TESS_files = []
periodogram_K2_files = []
periodogram_TESS_files = []
for file in os.listdir(lightcurve_directory):
    if ('_k2' in file):
        lightcurve_K2_files.append(file)
    if ('_tess' in file):
        lightcurve_TESS_files.append(file)
for file in os.listdir(periodogram_directory):
    if ('_k2' in file):
        periodogram_K2_files.append(file)
    if ('_tess' in file):
        periodogram_TESS_files.append(file)
```

```python
K2_data = {'Target': [],
           'lightcurve':[],
           'periodogram': [],
           }
TESS_data = {'Target': [],
             'lightcurve':[],
             'periodogram': [],
               }

for K2_i in range(len(lightcurve_K2_files)):
  targ_name = lightcurve_K2_files[K2_i].split('_k2')[0]
  lightcurve_temp_df = pd.read_csv(lightcurve_directory + lightcurve_K2_files[K2_i], delimiter=' ', names=['Time', 'Flux', 'Ferr'])
  periodogram_temp_df = pd.read_csv(periodogram_directory + periodogram_K2_files[K2_i], delimiter=' ', names=['Period', 'Power'])
  K2_data['Target'].append(targ_name)
  K2_data['lightcurve'].append(lightcurve_temp_df)
  K2_data['periodogram'].append(periodogram_temp_df)

for TESS_i in range(len(lightcurve_TESS_files)):
  targ_name = lightcurve_TESS_files[TESS_i].split('_tess')[0]
  lightcurve_temp_df = pd.read_csv(lightcurve_directory + lightcurve_TESS_files[TESS_i], delimiter=' ', names=['Time', 'Flux', 'Ferr'])
  periodogram_temp_df = pd.read_csv(periodogram_directory + periodogram_TESS_files[TESS_i], delimiter=' ', names=['Period', 'Power'])
  TESS_data['Target'].append(targ_name)
  TESS_data['lightcurve'].append(lightcurve_temp_df)
  TESS_data['periodogram'].append(periodogram_temp_df)
```
